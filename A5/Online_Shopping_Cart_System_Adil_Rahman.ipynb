## System Requirements
Each shopping cart must store:

- Customer name


- A list of item prices added to the cart


The system should be able to:

- Calculate total bill amount


- Apply discount rules


- Display cart details


- Combine two shopping carts


- Perform basic cart operations such as adding and removing items

## Part A – Class Creation (7 Marks)
Create a class named ShoppingCart with the following:

Attributes:
- customer_name → Name of the customer (string)


- items → List of item prices (list of numbers)


## Constructor:
- Use init() to initialize all attributes correctly


- If no items are provided, initialize with an empty list

## Part B – Core Methods Implementation (12 Marks)
Implement the following methods inside the ShoppingCart class:

## 1. add_item(price)
- Adds a new item price to the cart


- Price must be a positive number

## 2. remove_item(price)
- Removes one occurrence of the given price from the cart


- If the item does not exist, display a suitable message

## 3. calculate_total()
- Uses a loop or list comprehension


- Returns the total price of all items in the cart

## 4. apply_discount()
- Discount rule:



- Total ≥ 3000 → apply 10% discount


- Total < 3000 → no discount


- Returns the final payable amount

## Part C – Display & Utility Methods (6 Marks)
## 1. display_cart()
Prints the following in a well-formatted manner:

- Customer name


- Items in the cart


- Total amount


- Final amount after discount


- Number of items in the cart

## 2. clear_cart()
- Removes all items from the cart

- Displays a confirmation message

```python
class ShoppingCart:

  def __init__(self, customer_name, items=None):
    self.customer_name = customer_name

    if items is None:
      self.items = []
    else:
      self.items = items

  ## Add Item
  def add_item(self, price):
    if price > 0:
      self.items.append(price)
      print(f"Item with price {price} added to the cart.")
    else:
      print("Price must be a positive number.")

  ## Remove Item
  def remove_item(self, price):
    if price in self.items:
      self.items.remove(price)
      print(f"Item with price {price} removed from the cart.")
    else:
      print(f"Item with price {price} not found in the cart.")

  ## Calculate Total
  def calculate_total(self):
    total = sum([price for price in self.items])
    return total

  ## Apply Discount
  def apply_discount(self):
    total = self.calculate_total()

    if total >= 3000:
      final_payable = total * 0.9
    else:
      final_payable = total

    return final_payable

  # Combine two carts
  def combine_cart(self, other_cart):
    new_cart = ShoppingCart(self.customer_name + " & " + other_cart.customer_name)
    new_cart.items = self.items + other_cart.items
    return new_cart

  ## Display Cart
  def display_cart(self):
    print("-------------------------------------------------")
    print("Shopping Cart Details:\n")
    print(f"Customer Name: {self.customer_name}")
    print("Items in the Cart:")
    for item in self.items:
      print(f"- {item}")
    print(f"Total Amount: {self.calculate_total()}")
    print(f"Final Amount after Discount: {self.apply_discount()}")
    print(f"Number of Items in the Cart: {len(self.items)}")

    print("-------------------------------------------------")

  ## Clear Cart
  def clear_cart(self):
    self.items = []
    print("Shopping cart cleared.")

```

```python
cart1 = ShoppingCart("Adil")
cart2 = ShoppingCart("Rahman")

## Adding items and displaying cart
cart1.add_item(1000)
cart1.add_item(1500)
cart1.add_item(800)
cart1.display_cart()

## Removing items and displaying cart
cart1.remove_item(800)
cart1.display_cart()

## Another Cart
cart2.add_item(4000)
cart2.add_item(1000)
cart2.add_item(500)
cart2.display_cart()

## Removing items and displaying cart
cart1.remove_item(1000)
cart1.display_cart()


## Combine carts
combined = cart1.combine_cart(cart2)

combined.add_item(1000)
combined.display_cart()

combined.remove_item(500)
combined.display_cart()

print("\nCombined Cart")
combined.display_cart()

## Clearing All carts
cart1.clear_cart()
cart1.display_cart()

cart2.clear_cart()
cart2.display_cart()

combined.clear_cart()
combined.display_cart()
```
